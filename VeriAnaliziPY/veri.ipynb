```python
import pandas as pd

df = pd.read_csv("netflix_titles.csv")
print(df.head())
print(df.info())

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="type", palette="Set2")
plt.title("Netflix: Movie and TV Show Distribution")
plt.xlabel("İçerik Türü")
plt.ylabel("İçerik Sayısı")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.index, y=top_countries.values, palette="coolwarm")
plt.title("En Çok İçerik Üreten Ülkeler")
plt.xlabel("Ülke")
plt.ylabel("İçerik Sayısı")
plt.xticks(rotation=45)
plt.show()
```

```python
df = pd.read_csv("netflix_titles.csv")

top_directors = df['director'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_directors.values, y=top_directors.index, palette="viridis")
plt.title("Netflix'te en çok içerik üreten 10 yönetmen")
plt.xlabel("İçerik sayısı")
plt.ylabel("Yönetmenler")
plt.show()
```

```python
top_directors = df['director'].dropna().value_counts().head(10).index

df_top_directors = df[df['director'].isin(top_directors)][['director', 'title', 'type', 'release_year']]

print(df_top_directors.head(20))
```

```python
cast_data = df['cast'].dropna()

all_actors = cast_data.str.split(", ").explode().str.strip()

top_actors = all_actors.value_counts().head(10)
print(top_actors)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_actors.values, y=top_actors.index, palette="magma")

plt.title("Netflix'te en çok yer alan 10 oyuncu")
plt.xlabel("yer aldığı içerik sayısı")
plt.ylabel("Oyuncular")
plt.show()  
```

```python
df_non_india = df[df['country'] != 'India']

cast_data = df_non_india['cast'].dropna()
all_actors = cast_data.str.split(',').explode().str.strip()
top_actors = all_actors.value_counts().head(10)
print(top_actors)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x=top_actors.values,
    y=[actor for actor in top_actors.index if "Khan" not in actor],
    palette="magma"
)
plt.title("India dışındaki ülkelerde en çok yer alan 10 oyuncu")
plt.xlabel("yer aldığı içerik sayısı")
plt.ylabel("Oyuncular")
plt.show()
```

```python
#TR İÇİN
df_turkey = df[df['country']== "Turkey"]

cast_data = df_turkey['cast'].dropna()
all_actors = cast_data.str.split(',').explode().str.strip()

top_turkish_actors = all_actors.value_counts().head(15)

print(top_turkish_actors)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_turkish_actors.values, y=top_turkish_actors.index, palette='pastel')
plt.title("netflix türkiyede en çok yer alan oyuncular")
plt.xlabel("Yer aldığı içerik sayısı")
plt.ylabel("Oyuncular")
plt.show()
```
